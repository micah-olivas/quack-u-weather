## Evalulate model and build lookup table

First load duck data and weather data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load duck data
duck_df = pd.read_excel('/Users/micaholivas/Desktop/Personal/Programming Projects/quack-u-weather/Data/Duck_daily.xlsx')

# load weather data
weather_df = pd.read_csv('/Users/micaholivas/Desktop/Personal/Programming Projects/quack-u-weather/Data/3486317.csv')
```

```python
# coerce any columns to numeric that are not already
weather_df = weather_df.apply(pd.to_numeric, errors='ignore')

weather_df
```

```python
# get date range of duck data
duck_start_date = min(duck_df['Date'])
duck_end_date = max(duck_df['Date'])

# get range of data from weather data during duck data range
weather_df['DATE'] = pd.to_datetime(weather_df['DATE'])

# get all numeric columns from weather data
# weather_df = weather_df.select_dtypes(include=np.number)

# get min and max values for all columns
range_dict = {}
for col in weather_df.columns:
    if col != 'DATE':
        # if the column is numeric
        if np.issubdtype(weather_df[col].dtype, np.number):
            range_dict[col] = [weather_df[col].min(), weather_df[col].max()]

range_dict
```

```python
# now generate lists of values for each parameter
for col in range_dict.keys():
    range_dict[col] = np.linspace(range_dict[col][0], range_dict[col][1], num=10)
```
